# **MATPLOTLIB**

```
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
```

```
x=[1,2,3]
y=[6,5,6]
plt.plot(x,y)
plt.title('info')
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show()
```

```
xp=np.array([1,3])
yp=np.array([3,6])
plt.plot(xp,yp)
plt.show()
```

```
xp=np.array([1,3])
yp=np.array([3,6])
plt.plot(xp,yp,'o')
plt.show()
```

```
yp=np.array([1,3,5,6])
plt.plot(yp,marker="4")#differnt values have differnt shapes
plt.show()
```

```
plt.plot(yp,marker="o",ms=10,mec="b",mfc='y') #ms=marker size, mec=marker edge or border colour, mfc=marker facecolour
plt.show()
```

```
yp=np.array([1,3,5,6])
plt.plot(yp,linestyle='dotted')
plt.show()
```

```
yp=np.array([1,3,5,6])
plt.plot(yp,color="r")
```

```
yp=np.array([1,3,5,6])
plt.plot(yp,linewidth="4")
```

```
x=[1,2,3]
y=[6,5,6]
plt.subplot(1,2,1)#The layout is organized in rows and columns, which are represented by the first and second argument.The third argument represents the index of the current plot.
plt.plot(x,y)
x=[1,2,3]
y=[6,5,6]
plt.subplot(1,2,2)#the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x,y)
```

```
lan=["c++","java"]
valu=[25,35]
plt.bar(lan,valu)
```

```
plt.bar([0.5,1.5,2.5],[10,20,30],label="BMW",width=0.5)
plt.bar([0.1,1,2],[40,50,60],label="AUDI",color='r',width=0.5)
plt.legend()
plt.xlabel("days")
plt.ylabel("distance")
plt.show
```
